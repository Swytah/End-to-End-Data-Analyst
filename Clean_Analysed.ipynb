```python
#data load
import pandas as pd
df = pd.read_excel("sales_raw_500.xlsx")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df['Profit']=df['Sales']-df['Cost']
```

```python
#data column fix
df['Order_Date']=pd.to_datetime(df['Order_Date'])
```

```python
df.head(2)
```

```python
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month_name()
```

```python
df
```

```python
df.isna().sum()
```

```python
#analysis
```

```python
total_sales, total_cost, total_profit = (
    df['Sales'].sum(),
    df['Cost'].sum(),
    df['Profit'].sum()
)

print(f"Total Sales  : {total_sales:,}")
print(f"Total Cost   : {total_cost:,}")
print(f"Total Profit : {total_profit:,}")
```

```python
#Sales & Profit by Year
df.groupby('Year')[['Sales','Profit']].sum().reset_index()
```

```python
#Monthly sales trend
month_order = [
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
]

df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)
df.groupby('Month')['Sales'].sum().sort_index()
```

```python
#Region-wise Performance
df.groupby('Region')[['Sales','Profit']].sum().sort_values(by='Sales', ascending=False)
```

```python
'''Which region is most profitable?->North

Which region needs improvement?->East'''
```

```python
#Sales & Profit by Product
df.groupby('Product')[['Sales','Profit']].sum().sort_values(by='Profit', ascending=False)
```

```python
#Top customers by sales
df.groupby('Customer')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
df['Profit_Margin_%'] = (df['Profit'] / df['Sales']) * 100
#Average margin by product
df.groupby('Product')['Profit_Margin_%'].mean().sort_values(ascending=False)
```
